```python
def create_attack(attack_class, model, attack_name):
    try:
        if attack_name in ['PGD', 'FGSM', 'BIM', 'RFGSM', 'MIFGSM']:
            return attack_class(model, eps=8/255, alpha=2/255, steps=10)
        elif attack_name in ['CW']:
            return attack_class(model, c=1, lr=0.01, steps=50)
        elif attack_name in ['DeepFool']:
            return attack_class(model, steps=50)
        elif attack_name in ['AutoAttack']:
            return attack_class(model, norm='Linf', eps=8/255, version='standard')
        else:
            return attack_class(model)
    except Exception as e:
        print(f"Failed to create {attack_name}: {e}")
        return None
```

```python
from torchvision.datasets import ImageFolder
from torchvision.datasets.folder import default_loader, IMG_EXTENSIONS
import os

class SafeImageFolder(ImageFolder):
    def __init__(self, root, transform=None, target_transform=None, loader=default_loader, is_valid_file=None):

        classes, class_to_idx = self.find_classes_safe(root)
        samples = self.make_dataset_safe(root, class_to_idx, IMG_EXTENSIONS, is_valid_file)
        if len(samples) == 0:
            raise RuntimeError(f"Found 0 valid image files in {root}.")

        self.loader = loader
        self.extensions = IMG_EXTENSIONS
        self.root = root
        self.transform = transform
        self.target_transform = target_transform
        self.classes = classes
        self.class_to_idx = class_to_idx
        self.samples = samples
        self.targets = [s[1] for s in samples]

    def __getitem__(self, index):
        path, target = self.samples[index]
        sample = self.loader(path)
        if self.transform is not None:
            sample = self.transform(sample)
        if self.target_transform is not None:
            target = self.target_transform(target)
        return sample, target, path

    def find_classes_safe(self, dir):
        """Find only class folders that are non-empty"""
        classes = []
        class_to_idx = {}
        for entry in sorted(os.listdir(dir)):
            path = os.path.join(dir, entry)
            if os.path.isdir(path):
                valid = False
                for fname in os.listdir(path):
                    if fname.lower().endswith(tuple(e.lower() for e in IMG_EXTENSIONS)):
                        valid = True
                        break
                if valid:
                    classes.append(entry)
        class_to_idx = {cls_name: i for i, cls_name in enumerate(classes)}
        return classes, class_to_idx

    def make_dataset_safe(self, directory, class_to_idx, extensions, is_valid_file):
        """Build the dataset, skipping folders with no valid images"""
        instances = []
        for target_class in sorted(class_to_idx.keys()):
            class_index = class_to_idx[target_class]
            target_dir = os.path.join(directory, target_class)
            if not os.path.isdir(target_dir):
                continue
            for root, _, fnames in sorted(os.walk(target_dir)):
                for fname in sorted(fnames):
                    path = os.path.join(root, fname)
                    if fname.lower().endswith(tuple(ext.lower() for ext in extensions)):
                        item = path, class_index
                        instances.append(item)
        return instances
```

```python
import os
import torch
import timm
import torchattacks
import pandas as pd
from tqdm import tqdm
from PIL import Image
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from timm.data import resolve_data_config
from timm.data.transforms_factory import create_transform
import torchvision.transforms.functional as TF

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

model = timm.create_model('vit_base_patch16_224', pretrained=True)
model = model.to(device).eval()
config = resolve_data_config({}, model=model)
transform = create_transform(**config)

def safe_loader(path):
    try:
        with open(path, 'rb') as f:
            return Image.open(f).convert('RGB')
    except:
        return None

class SafeImageFolder(datasets.ImageFolder):
    def __init__(self, root, transform):
        super().__init__(root, transform=transform, loader=safe_loader)

    def __getitem__(self, index):
        path, target = self.samples[index]
        sample = self.loader(path)

        while sample is None:
            index = (index + 1) % len(self.samples)
            path, target = self.samples[index]
            sample = self.loader(path)

        if self.transform is not None:
            sample = self.transform(sample)
        return sample, target, path

dataset = SafeImageFolder('/content/drive/MyDrive/my231n/imagenet_val_split', transform=transform)
test_loader = DataLoader(dataset, batch_size=16, shuffle=False)
print(f"Total batches: {len(test_loader)}")

def save_adversarial_images(images, labels, paths, attack_name, output_root, metadata):
    attack_dir = os.path.join(output_root, attack_name)
    os.makedirs(attack_dir, exist_ok=True)

    for img_tensor, label, img_path in zip(images, labels, paths):
        class_name = os.path.basename(os.path.dirname(img_path))
        filename = os.path.basename(img_path)
        save_path = os.path.join(attack_dir, f"{class_name}_{filename}")

        try:
            img = TF.to_pil_image(img_tensor.cpu().clamp(0, 1))
            img.save(save_path)

            metadata.append({
                'image_path': save_path,
                'original_class': class_name,
                'attack_type': attack_name
            })
        except Exception as e:
          print(f"Error saving image: {e}")
            continue

def generate_adversarial_dataset(model, attack_class, attack_name, loader, output_root):
    attack = attack_class(model)
    metadata = []

    for images, labels, paths in tqdm(loader, desc=f"Generating {attack_name}"):
        images, labels = images.to(device), labels.to(device)

        try:
            images_input = images.clone().detach().requires_grad_(True)
            adv_images = attack(images_input, labels)
            save_adversarial_images(adv_images, labels, paths, attack_name, output_root, metadata)
        except Exception as e:
            continue

        if torch.cuda.is_available():
            torch.cuda.empty_cache()

    return metadata

output_root = '/content/drive/MyDrive/my231n/adversarial_val_dataset'
all_metadata = []

attacks = [
    ('FGSM', torchattacks.FGSM),
    ('PGD', torchattacks.PGD),
    ('BIM', torchattacks.BIM),
    ('CW', torchattacks.CW),
    ('EOTPGD', torchattacks.EOTPGD),
    ('APGD', torchattacks.APGD),
    ('APGDT', torchattacks.APGDT),
    ('FAB', torchattacks.FAB),
    ('VANILA', torchattacks.VANILA),
    ('GN', torchattacks.GN),
    ('DeepFool', torchattacks.DeepFool),

]

for name, cls in attacks:
    meta = generate_adversarial_dataset(model, cls, name, test_loader, output_root)
    all_metadata.extend(meta)

df_meta = pd.DataFrame(all_metadata)
df_meta.to_csv(os.path.join(output_root, 'metadata.csv'), index=False)
print(f" Saved {len(all_metadata)} adversarial images.")
```

```python
import torchvision.transforms.functional as TF
from tqdm import tqdm
import os

def save_clean_images(loader, output_root, existing_meta=None):
    clean_dir = os.path.join(output_root, 'clean')
    os.makedirs(clean_dir, exist_ok=True)
    metadata = [] if existing_meta is None else existing_meta

    for images, labels, paths in tqdm(loader, desc="Saving clean images"):
        for img_tensor, label, img_path in zip(images, labels, paths):
            class_name = os.path.basename(os.path.dirname(img_path))
            filename = os.path.basename(img_path)
            save_path = os.path.join(clean_dir, f"{class_name}_{filename}")


            img = TF.to_pil_image(img_tensor.cpu().clamp(0, 1))
            img.save(save_path)


            metadata.append({
                'image_path': save_path,
                'original_class': class_name,
                'attack_type': 'clean'
            })

    return metadata
```

```python
import pandas as pd
import torch
from torchvision import transforms
from torchvision.datasets.folder import default_loader
from tqdm import tqdm
import timm
from collections import defaultdict


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vit = timm.create_model('vit_base_patch16_224', pretrained=True).eval().to(device)

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=(0.5, 0.5, 0.5), std=(0.5, 0.5, 0.5))
])


metadata_path = "/content/drive/MyDrive/my231n/adversarial_val_dataset/metadata_with_clean.csv"
class_map_path = "/content/drive/MyDrive/my231n/cs231n_project/class_map.csv"

df = pd.read_csv(metadata_path)
class_map_df = pd.read_csv(class_map_path, header=None, names=["class_name", "wnid", "class_id"])

name_to_id = dict(zip(class_map_df["class_name"], class_map_df["class_id"]))

acc_by_attack = defaultdict(lambda: {"correct": 0, "total": 0})

for _, row in tqdm(df.iterrows(), total=len(df)):
    img_path = row["image_path"]
    label_name = row["original_class"]
    attack_type = row["attack_type"]

    if label_name not in name_to_id:
        print(f"Warning: '{label_name}' not found in class map, skipping.")
        continue

    gt_label = name_to_id[label_name]

    try:
        img = default_loader(img_path)
        img = transform(img).unsqueeze(0).to(device)

        with torch.no_grad():
            pred = vit(img).argmax(dim=1).item()

        acc_by_attack[attack_type]["total"] += 1
        acc_by_attack[attack_type]["correct"] += int(pred == gt_label)

    except Exception as e:
        print(f"Error on {img_path}: {e}")
        continue

for attack, stats in acc_by_attack.items():
    acc = stats["correct"] / stats["total"] if stats["total"] > 0 else 0.0
    print(f"{attack:10s}: {acc:.4f} ({stats['correct']} / {stats['total']})")
```
